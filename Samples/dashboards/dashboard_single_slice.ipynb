```python
import os,sys
import OpenVisus  as  ov
from  OpenVisus.dashboards  import  Slices
from bokeh.io import output_notebook, show
from bokeh.application import Application
from bokeh.application.handlers import FunctionHandler

os.environ["BOKEH_ALLOW_WS_ORIGIN"]="localhost:8888"
output_notebook()
```

```python

# change as needed
db=ov.LoadDataset(f"https://s3.us-west-1.wasabisys.com/Pania_2021Q3_in_situ_data/workflow/fly_scan_id_112509.h5/r/idx/1mb/visus.idx?profile=wasabi&cached=idx&blob_extension=.bin.zz")
slice=Slices()
slice.setPalette("Greys256", palette_range=(0,65535))
slice.setDataset(db, layout=2)

def modify_doc(doc):
    doc.theme = 'caliber'
    doc.add_root(slice.layout)
    doc.add_periodic_callback(slice.getCentralLayout, 100)
   
```
