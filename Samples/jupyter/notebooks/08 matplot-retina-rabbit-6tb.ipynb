# Import matplotlib

```python
import os,sys
import matplotlib.pyplot as plt, skimage
%matplotlib notebook

def ShowData(data,extent):
    fig, ax = plt.subplots()
    im = ax.imshow(data, extent=extent) 
    plt.colorbar(im)
    return fig,im,ax
```

# Import OpenVisus

```python
sys.path.append(r"C:\projects\OpenVisus\build\RelWithDebInfo")
import OpenVisus as  ov
from OpenVisus.dashboards import Slice,Slices
from OpenVisus.pyquery import PyQuery
os.environ["VISUS_NETSERVICE_VERBOSE"]="0"
print("OpenVisus imported")
```

# Load Dataset

```python
url='http://atlantis.sci.utah.edu/mod_visus?dataset=rabbit&cached=1'
db=ov.LoadDataset(url)
print(f"Loaded dataset \n\t url={db.getUrl()} \n\t fields={db.getFields()} \n\t logic_box={db.getLogicBox()}")
W,H=db.getLogicBox()[1]

def ReadData(cx, cy, zoom, quality=0, verbose=True): 
    x1,x2 = int(cx - zoom*W*0.5), int(cx + zoom*W*0.5)
    y1,y2 = int(cy - zoom*H*0.5), int(cy + zoom*H*0.5)  
    aligned_box, __delta, __num_pixels = PyQuery.getAlignedBox(db,[[x1,y1],[x2,y2]],db.getMaxResolution()+quality)
    (x1,y1),(x2,y2)=aligned_box
    extent=[aligned_box[0][0],aligned_box[1][0],aligned_box[0][1],aligned_box[1][1]]   
    data=db.read(logic_box=aligned_box, time=1,quality = quality)
    if verbose: print(f"data.shape={data.shape} extent={extent}")
    return data,extent
```

# Display a single XY slice

```python
fig, plots = plt.subplots(nrows=1, ncols=1)
data,extent = ReadData(W//2,H//2,quality = -12, zoom=1.0)
ShowData(data,extent)
plt.show() 
```

# Mark the area of interest and show it

```python
piece_data, piece_extent = ReadData(W//2,H//2,zoom=0.1,quality=-12)

ShowData(data,extent)

x1,x2,y1,y2=piece_extent
plt.plot([x1,x1],[ 0, H], color='r')
plt.plot([x2,x2],[ 0, H], color='r')
plt.plot([ 0, W],[y1,y1], color='g')  
plt.plot([ 0, W],[y2,y2], color='g')  
plt.show()

ShowData(piece_data,piece_extent)
plt.show()
```
